# Data Transformation

```python
def transform_and_append(data, csv):
    csv['Date'] = pd.to_datetime(csv['Date'], dayfirst = True)
    csv.columns = ['Date', 'Dernier', 'Ouverture', 'Plus Haut', 'Plus Bas', 'Volume', 'Variation %', 'Societe']
    for col in csv.columns[1:-1] :
        if csv[col].dtype != (float or int):
            csv[col] = csv[col].str.replace("\\.|%","", regex = True)
            csv[col] = csv[col].str.replace(",",".", regex = False)
            if col != 'Volume':
                csv[col] = pd.to_numeric(csv[col])
    
    csv['Volume'] = csv['Volume'].replace(r'[KM]+$', '', regex=True).astype(float) * csv['Volume'].str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int)

    data.append(csv)
```

```python
files = [filename for filename in glob.glob("../StockMarketData/*.csv")]
data = []
for file in files:
    csv = pd.read_csv(file)
    csv['societe'] = file[19:22]
    transform_and_append(data, csv)
data = pd.concat(data, axis=0).sort_values(by=['Date'], ascending = True).set_index('Date')

data
```

# Arima

```python
# Arima is a univariate model

targets = {}

for index, Companie in enumerate(Companies):
    targets[Companies_Title[index]] = Companie.Dernier
```

### Bulduing an Arima Model and chosing params manually 

```python
def is_stationary(ts):

    result = adfuller(ts)

    if (result[1] <= 0.05) & (result[4]['5%'] > result[0]):
        return True
    else:
        return False
```

```python
def find_differencing_order(ts):
    diff = ts.copy()
    d = 0
    while not is_stationary(diff) :
        diff = np.diff(diff, n=1)
        d += 1
    return d
```

```python
#find the differencing order (d)

d_order = {}
for Companie_title, target in targets.items():
    d = find_differencing_order(target)
    print('The order of differencing (d) for ', Companie_title,'is : ', d)
    d_order[Companie_title] = d
```

```python
for Companie_title, target in targets.items():
    d = d_order[Companie_title]
    
    fig, ax = plt.subplots(1,2, figsize=(12, 3))
    
    plot_acf(np.diff(target, n=d), ax = ax[0])
    ax[0].set_title('ACF : '+ Companie_title)
    plot_pacf(np.diff(target, n=d), ax = ax[1], method='ywm')
    ax[1].set_title('PACF : '+ Companie_title)
```

```python
ARIMA_Params = {}

ARIMA_Params['BCP'] = (1,1,1) #{'p' : 1, 'd' : 1, 'q' : 1}
ARIMA_Params['CDM'] = (2,1,1) 
ARIMA_Params['BOA'] = (3,1,3)
ARIMA_Params['ATW'] = (1,1,1)
```

```python
#train_test split
train_data, test_data = {}, {}

for Companie_title, target in targets.items():
    train_len = int(len(targets[Companie_title]) * 0.75)
    train_data[Companie_title] = target[:train_len]
    test_data[Companie_title] = target[train_len:]
```

```python
# Build Model

def Arima_model(companie):
    
    train = train_data[companie]
    test = test_data[companie]

    model = ARIMA(train, order = ARIMA_Params[companie]).fit()
     
    fcast_res1 = model.get_forecast(len(test)).summary_frame(alpha=0.10)
    
    fc = fcast_res1['mean']
    se = fcast_res1['mean_se']
    conf_lower = fcast_res1['mean_ci_lower']
    conf_upper = fcast_res1['mean_ci_upper']


    # Make as pandas series
    fc_series = pd.Series(fc.values, index=test.index)
    lower_series = pd.Series(conf_lower.values, index=test.index)
    upper_series = pd.Series(conf_upper.values, index=test.index)
    
    return model, fc_series, lower_series, upper_series

def plot_forcast_vs_actual(fc_series, lower_series, upper_series, company):
    train = train_data[company]
    test = test_data[company]
    
    plt.figure(figsize=(8,3), dpi=100)
    plt.plot(train, label='training')
    plt.plot(test, label='actual')
    plt.plot(fc_series, label='forecast')
    plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.15)
    plt.title('Forecast vs Actuals : ' + company)
    plt.legend(loc='upper left', fontsize=8)
    plt.show()
```

```python
models = {}
for Companie in Companies_Title:
    models[Companie], fc_series, lower_series, upper_series = Arima_model(Companie)
    plot_forcast_vs_actual(fc_series, lower_series, upper_series, Companie)
```

### reshaping the data

```python
def Shape_data(data, steps):
    features_num = len(data.columns) 
    data = data.values
    steps_len = steps + 1 
    result = []
    
    for index in range(len(data) - steps_len):
        result.append(data[index: index + steps_len])
    
    result = np.array(result)
    a = round(0.9 * result.shape[0])
    train = result[:int(a), :]  
    
    x_train = train[:, :-1] 
    y_train = train[:, -1][:,-1]
    
    x_test = result[int(a):, :-1] 
    y_test = result[int(a):, -1][:,-1]

    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], features_num))
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], features_num))  

    return [x_train, y_train, x_test, y_test]

```

### Model Building 

```python

def model_Building(data):
    model = Sequential()
    model.add(LSTM(256, input_shape=(data[1], data[0]), return_sequences=False))        
    
    #model.add(LSTM(256, input_shape=(data[1], data[0]), return_sequences=False))        
    #model.add(LSTM(64, input_shape=(data[1], data[0]), return_sequences=False))        
   
    model.add(Dense(32,activation='relu'))        
    
    model.add(Dense(1,activation='relu'))
            
    start = time.time()
    model.compile(loss='mse',optimizer='adam', metrics=['MeanSquaredError'])
    print("Compilation Time : ", time.time() - start)
    return model
```

```python
Steps = 20
```

### Model Evaluation

```python
def evaluation(model, X_train, y_train, X_test, y_test):
    tr_Score = model.evaluate(X_train, y_train, verbose=0)
    print('Train : %.5f MSE (%.2f RMSE)' % (tr_Score[0], math.sqrt(tr_Score[0])))

    ts_Score = model.evaluate(X_test, y_test, verbose=0)
    print('Test : %.5f MSE (%.2f RMSE)' % (ts_Score[0], math.sqrt(ts_Score[0])))
    return tr_Score[0], ts_Score[0]
```

### Data denormalization 

```python

def denormalization(df, normalized_data): 
    df = df['Dernier'].values.reshape(-1,1)
    normalized_data = normalized_data.reshape(-1,1)
    print(df.shape)
    print(normalized_data.shape)
    min_max_scaler = MinMaxScaler()
    a = min_max_scaler.fit_transform(df)
    new = min_max_scaler.inverse_transform(normalized_data)
    return new


```
